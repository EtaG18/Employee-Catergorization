```python
import pandas as pd
import numpy as np
import string
string.punctuation
import nltk
from nltk.tokenize import wordpunct_tokenize
import re
import nltk
nltk.download('stopwords')


#Printing the data
data=pd.read_csv('/content/drive/MyDrive/IIIT/Thesis/SMS_Spam_Data/spam.csv',encoding='latin-1')
print(data)

#We will drop the last three columns as they are of no use to us and create a new dataframe called new_data
new_data=data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
print(new_data)

#Printing the shape of the data to understand the number of samples and the number of features in the data
new_data.shape

#Finding out the number of samples present in the ham and spam classes
new_data['v1'].value_counts()
```

```python
#Now we are applying the tokenization and removing punctuation from the sample sentence to understand how it works

sample_data='I am, a very happy person, if you area $also% & happy *! clap your hands!!'

def remove_punctuation(text):
  punctuationfree="".join([i for i in text if i not in string.punctuation])
  return punctuationfree

cleaned_sample_data=remove_punctuation(sample_data)
print('Sentence with punctuation marks removed ---->',cleaned_sample_data)

sample_data_list=sample_data.split()
tokenized_data=wordpunct_tokenize(sample_data)
print('The actual sentence---->',sample_data)
print('The sentence which is tokenized---->',tokenized_data)
```

```python
#storing the puntuation free text from the complete dataframe

new_data['clean_msg']= new_data['v2'].apply(lambda x:remove_punctuation(x))
new_data.head()

#defining function for tokenization

def tokenization(text):
    tokens = re.split('W+',text)
    return tokens
#applying function to the column
new_data['msg_tokenied']= new_data['clean_msg'].apply(lambda x: tokenization(x))

#Printing the tokenized data
print(new_data.head())
```

```python
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're"]

#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output

#applying the function
new_data['no_stopwords']= new_data['msg_tokenied'].apply(lambda x:remove_stopwords(x))
print(new_data)
```
