```python
sex_map = {'Male':1, 'Female':0}
data['Gender'] = data['Gender'].map(sex_map)

pyth_map = {'Yes':1, 'No':0}
data['Python_exp'] = data['Python_exp'].map(pyth_map)

edu_map = {'Graduate':1, 'Not Graduate':0}
data['Education'] = data['Education'].map(edu_map)

intern_map = {'Yes':1, 'No':0}
data['Internship'] = data['Internship'].map(intern_map)

locat_map = {'Urban':3, 'Semiurban':2, 'Rural':1}
data['Location'] = data['Location'].map(locat_map)

status_map = {'Y':1, 'N':0}
data['Recruitment_Status'] = data['Recruitment_Status'].map(status_map)


data.head()
```

```python
import numpy as np

data['Gender'].fillna(0, inplace=True)

data['Experience_Years'].fillna(0, inplace=True)

data['Python_exp'].fillna(0, inplace=True)

data['Offer_History'].fillna(1, inplace=True)

data['Internship'].fillna(np.mean(data['Internship']), inplace=True)

data['Salary * 10E4'].fillna(np.mean(data['Salary * 10E4']), inplace=True)
```

```python
m = np.random.rand(len(data))<0.72

train = data[m]
test = data[~m]
```

```python
x_train = np.asarray(train[['Gender', 'Python_exp', 'Experience_Years', 'Education', 'Internship', 'Score', 'Salary * 10E4', 'Offer_History']])
y_train = np.asarray(train[['Recruitment_Status']])

x_test = np.asarray(test[['Gender', 'Python_exp', 'Experience_Years', 'Education', 'Internship', 'Score', 'Salary * 10E4', 'Offer_History']])
y_test = np.asarray(test[['Recruitment_Status']])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

#Logistic Regression model
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression() #Define
log_model.fit(x_train, y_train) #fit
predict = log_model.predict(x_test) #predict

print("Train data Accuracy :", log_model.score(x_train, y_train))
print("Test data Accuracy :", log_model.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
import plotly.express as px

model=RandomForestClassifier()

model.fit(x_train,y_train)

y_pred=model.predict(x_test)

print("Training Accuracy :", model.score(x_train, y_train))
print("Testing Accuracy:", model.score(x_test, y_test))

# cm = confusion_matrix(y_test, y_pred)
# fig = px.imshow(cm,text_auto=True, title='Confusion Matrix of the Model')
# fig.update_layout(xaxis_title='Actual', yaxis_title='Predicted')
```
